```python
import numpy as np
import itertools
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.randint(2)
```

```python
x = range(0,10)
y = range(0,10)
data_xy = list(itertools.product(*(x,y)))
```

```python
data_xyv = [e + (np.random.randint(2),) for e in data_xy]
```

```python
df = pd.DataFrame(np.array(data_xyv),columns=['x1','x2','y'])
df.head()
```

```python
plt.scatter(df.x1,df.x2,c=df.y)
```

```python
def jitter(data,stdev):
    N = len(data)
    return data + np.random.randn(N)*stdev
sigma = 200 
plt.scatter(jitter(df.x1,sigma),jitter(df.x2,sigma) ,c=df.y)
```
